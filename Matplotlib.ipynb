```python
#1. Draw a line in a diagram from position (1, 3) to (2, 10) then to (6, 12) and finally to position (18, 20). (Mark each point with a beautiful green colour and set line colour to red and line style dotted).
import numpy as np 
import matplotlib.pyplot as plt 
x=np.array([1,2,6,8]) 
y=np.array([3,8,1,10]) 
plt.plot(x,y,marker='o',mec='g',mfc='g',linestyle="dashed",color="red") 
plt.show 
```

```python
''''2.Draw a plot for the following data:
Temperature in degree Celsius Sales
 12 100
 14 200
 16 250
 18 400
 20 300
 22 450
 24 500'''
import numpy as np 
import matplotlib.pyplot as plt 
x=np.array([12,14,16,18,20,22]) 
y=np.array([100,200,300,400,500,600]) 
plt.plot(x,y) 
plt.title("Temperature in degree Celsius v/s Sales ") 
plt.xlabel("Temperature in degree Celsius ") 
plt.ylabel("Sales") 
plt.show() 
```

```python
'''5.Aim: Consider the following data. 
 Programming languages: Java Python PHP JavaScript C# C++ 
 Popularity : 22.2 17.6 8.8 8 7.7 6.7 
(i) Write a Python programming to display a bar chart of the popularity of programming 
Languages. 
(ii) Write a Python programming to display a horizontal bar chart of the popularity of 
programming Languages(Give Red colour to the barchart ) 
 (iii) Write a Python programming to display a bar chart of the popularity of 
programming Languages. Use different color for each bar.
(iv). Write a Python programming to display a pie chart of the popularity of 
programming Languages.'''
import matplotlib.pyplot as plt 
import numpy as np 
a=np.array(["java","python","php","JavaScript","C#","C++"]) 
b=np.array([22.2,17.6,8.8,8,7.7,6.7]) 
plt.title("Popularity v/s programming language") 
plt.xlabel("Programming lang") 
plt.ylabel("poplularity") 
plt.bar(a,b) 
plt.show() 
plt.title("Popularity v/s programming language") 
plt.barh(a,b,color="red") 
plt.show() 
plt.title("Popularity v/s programming language") 
plt.bar(a,b,color=["red","blue","green","pink","yellow"]) 
plt.show() 
plt.title("Popularity v/s programming language") 
la=["java","python","php","JavaScript","C#","C++"] 
plt.pie(b,labels=la) 
plt.show() 
```
