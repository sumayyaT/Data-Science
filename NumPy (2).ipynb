```python
#1.Program to multiply two given arrays of same size element-by-element.
import numpy as np 
num1=([[2,4,5],[2,6,1]]) 
num2=([[3,4,7,],[2,6,1]]) 
print("The first array is") 
print(num1) 
print("The second array is") 
print(num2) 
num3=np.multiply(num1,num2) 
print("After multiplication:") 
print(num3) 
```

```python
#2.Program to create an element-wise comparison (greater, greater_equal, lesseer).
import numpy as np 
num1=([[2,4,5],[2,6,1]]) 
num2=([[3,4,7,],[2,6,1]]) 
print("First Array") 
print(num1) 
print("Second array") 
print(num2) 
2 
print("Comparison : greater") 
print(np.greater(num1,num2)) 
print("Comparison : greater_equal") 
print(np.greater_equal(num1,num2)) 
print("Comparison : lesser") 
print(np.less(num1,num2)) 
print("Comparison : less_equal") 
print(np.less_equal(num1,num2)) 
print("Comparison : equal") 
print(np.equal(num1,num2)) 
```

```python
#7.Program to compute sum of all elements, sum of each column and sum of each row
import numpy as np 
a=np.array([[1,2,3],[1,2,3]]) 
print("Sum of all elements") 
print(np.sum(a)) 
print("Sum of each column") 
print(np.sum(a,axis=0)) 
print("Sum of each row") 
print(np.sum(a,axis=1)) 
```

```python
#8.Program to save a given array to a text file and load it. 
import numpy as np 
a=np.arange(12).reshape(4,3) 
print("Array:") 
print(a) 
np.savetxt("temp.txt",a) 
res=np.loadtxt("temp.txt") 
print("Array after converted to .txt") 
print(res) 
```
