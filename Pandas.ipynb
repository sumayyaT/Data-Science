```python
#6.Given a dataframe, sort it by multiple columns. 
import pandas as pd 
dic = {'name': ['gopi', 'Kuttappu', 'Archana'], 
 'age' : [22, 25, 24], 
 'occ' : ['engineer', 'doctor', 'accountant']} 
df = pd.DataFrame(dic) 
print("Data frame") 
print(df) 
print("After sorting data frame by name and occ") 
d_sorted =df.sort_values(by = ['name', 'occ']) 
print(d_sorted) 
```

```python
#7. Given a dataframe with custom indexing, convert and it to default indexing and display it. 
import pandas as pd 
dic = {'id':[1,2,3], 
 'name': ['gopi', 'Kuttappu', 'Archana'], 
 'age' : [22, 25, 24], 
 'occ' : ['engineer', 'doctor', 'accountant'], 
 } 
df = pd.DataFrame(dic) 
print("Data frame") 
print(df) 
print("After setting default index as id") 
index=df.set_index('id') 
print(index) 
print("After setting default index as id") 
reindex=index.reset_index() 
print(reindex) 
```

```python
#9. Given is a dataframe showing name, occupation,salary of people. Find the average salary per occupation. 
import pandas as pd 
dic = { 
 'name': ['gopi', 'Kuttappu', 'Archana'], 
 'salary' : [75000, 50000, 100000], 
 'occ' : ['engineer', 'doctor', 'accountant'], 
 } 
df = pd.DataFrame(dic) 
print("Data frame") 
print(df) 
avg = df.groupby('occ')['salary'].mean() 
print("average salary per occupation") 
print(avg) 
```

```python
#10. Given is a dataframe showing Company Names (cname) and corresponding Profits (profit). Convert the values of Profit column such that values in it greater than 0 are set to True and the rest are set to False. 
import pandas as pd 
dic = { 
 'c_name': ['Tcs', 'Wipro', 'Microsoft'], 
 'profit' : [200000, -2500000, 24000000], 
 } 
df = pd.DataFrame(dic) 
print("Dataframe") 
print(df) 
print("After profit column to values True when profit greater than otherwise False.") 
df['profit'] = df['profit'].apply(lambda x:x>0) 
print(df) 
```
